```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Sales.csv")
```

```python
# 1. Missing Values
print("Missing values per column:")
print(df.isnull().sum())
```

```python

# 2. Duplicates
duplicate_rows = df.duplicated().sum()
print("Number of duplicate rows",duplicate_rows)
if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)

# Convert date columns to datetime objects
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

# Handle missing Postal Code values
df['Postal Code'].fillna('unknown', inplace=True)
print("Missing 'Postal Code' values filled with unknown.")

# Extract Order Month(derived fearures)
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year
```

```python
##active vs inactive based oon purchase in the last 90 days
df.describe()
```

```python
customer_profile = df.groupby("Customer ID").agg(
    First_Purchase_Month=("Order Month", "min"),
    Total_Orders=("Order ID", "nunique"))

# join this with the dateframe dataframe
df = df.join(customer_profile, on="Customer ID")

# Flag
df["Is_Repeat_Customer"] = df["Total_Orders"] > 1


#  Mapping (New vs Churned vs Active)
all_months = df["Order Month"].unique()
all_customers = df["Customer ID"].unique()

# Customer-Month combinations grid
grid = pd.MultiIndex.from_product(
    [all_customers, all_months], names=["Customer ID", "Order Month"]
).to_frame()
grid.reset_index(drop=True, inplace=True)

# Determine if a customer actually purchased in that specific month
monthly_purchases = (
    df.groupby(["Customer ID", "Order Month"]).size().to_frame("Purchase_Count")
)
grid = grid.join(monthly_purchases, on=["Customer ID", "Order Month"])
grid["Purchased_This_Month"] = grid["Purchase_Count"].notna()

# Pull (first purchase) month for each customer
grid = grid.join(customer_profile["First_Purchase_Month"], on="Customer ID")

# Filter grid to only check months
grid = grid[grid["Order Month"] >= grid["First_Purchase_Month"]].copy()
grid.sort_values(by=["Customer ID", "Order Month"], inplace=True)

# Calculate tracking shifts for historical activity
grid["Purchased_Prev_Month"] = grid.groupby("Customer ID")[
    "Purchased_This_Month"
].shift(1).fillna(False)


# Flags:
grid["Status_New"] = grid["Order Month"] == grid["First_Purchase_Month"]
grid["Status_Active"] = grid["Purchased_This_Month"]
grid["Status_Churned"] = grid["Purchased_Prev_Month"] & ( ~grid["Purchased_This_Month"])


# 3. Rolling 90-Day Active vs Inactive State
max_date = df["Order Date"].max()

last_90_days_customers = df[df["Order Date"] >= (max_date - pd.Timedelta(days=90))][
    "Customer ID"
].unique()

df["Status_90D_Active"] = df["Customer ID"].isin(last_90_days_customers)
df["Status_90D_Inactive"] = ~df["Status_90D_Active"]

print(" Sample of Customer Monthly Lifecycle Transitions ")
print(grid[["Customer ID","Order Month","Status_New", "Status_Active", "Status_Churned",] ].head(10)
)
```

```python
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
 #STEP 3: ML CUSTOMER SEGMENTATION (RFM K-MEANS)
snapshot_date = df["Order Date"].max() + pd.Timedelta(days=1)

rfm = df.groupby("Customer ID").agg(
    Recency=("Order Date", lambda x: (snapshot_date - x.max()).days),
    Frequency=("Order ID", "nunique"),
    Monetary=("Sales", "sum"),
)

# Preprocessing for K-Means (Log shift + scaling)
rfm_scaled = StandardScaler().fit_transform(np.log1p(rfm))
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
rfm["Cluster_ID"] = kmeans.fit_predict(rfm_scaled)

# Map human-readable
def heuristic_namer(row):
    if row["Recency"] <= rfm["Recency"].quantile(0.3) and row["Frequency"] >= rfm["Frequency"].median():
        return "Champions"
    elif row["Recency"] > rfm["Recency"].quantile(0.7):
        return "Slipping / Defecting"
    else:
        return "Mid-Market Mid-Tier"

rfm["Customer_Segment"] = rfm.apply(heuristic_namer, axis=1)

# STEP 4: CHURN PREDICTION

max_date = df["Order Date"].max()
churn_cutoff_days = 90

ml_features = df.groupby("Customer ID").agg(
    Total_Sales=("Sales", "sum"),
    Total_Quantity=("Quantity", "sum"),
    Avg_Discount=("Discount", "mean"),
    Total_Orders=("Order ID", "nunique"),
    Avg_Profitability=("Profit", "mean"),
    Days_Since_Last_Order=("Order Date", lambda x: (max_date - x.max()).days),
)

# Label: 1 if not purchased in last 90 days, else 0
ml_features["Target_Churned"] = (ml_features["Days_Since_Last_Order"] > churn_cutoff_days).astype(int)

# Split and isolate inputs
X = ml_features[["Total_Sales", "Total_Quantity", "Avg_Discount", "Total_Orders", "Avg_Profitability"]]
y = ml_features["Target_Churned"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
rf_model = RandomForestClassifier(n_estimators=150, class_weight="balanced", random_state=42)
rf_model.fit(X_train, y_train)

# Evaluate Prediction Performance
y_pred = rf_model.predict(X_test)
print(classification_report(y_test, y_pred))
print(f"ROC-AUC Score Accuracy: {roc_auc_score(y_test, rf_model.predict_proba(X_test)[:,1]):.2f}")

ml_features["Churn_Probability"] = rf_model.predict_proba(X)[:, 1]

# STEP 5: EXPORT TABLES FOR POWER BI

print("\nExporting processed structured datasets for Power BI integration...")

# Flat Order Book augmented with customer traits
# The 'Total_Orders' column is already present in 'df' from a previous join with 'customer_profile'.
# Therefore, we can directly use 'df' and add the 'Is_Repeat_Customer' column.
bi_orders = df.copy()
bi_orders["Is_Repeat_Customer"] = bi_orders["Total_Orders"] > 1
bi_orders.to_csv("pbi_orders_fact.csv", index=False)

# Monthly Cohort State Table
grid.to_csv("pbi_monthly_cohort_matrix.csv", index=False)

# Customer Segment Master Mapping File
customer_master = rfm[["Customer_Segment"]].join(ml_features[["Churn_Probability", "Target_Churned"]])
customer_master.reset_index().to_csv("pbi_customer_dim.csv", index=False)
```
